##Gráficos

```python
#Sessão matplotlib

import matplotlib.pyplot as plt
import numpy as np

```

```python
##prática 1
xa_array= np.linspace(0,2*np.pi,500)
ya_array = np.sin(xa_array)

xb_array= np.linspace(0,2*np.pi,500)
yb_array = np.sin(xb_array + np.pi/2)


plt.plot(xa_array,ya_array,'r-')
plt.plot(xb_array,yb_array, 'g-')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

```

```python
##prática 2

fig, ax= plt.subplots()

ax.plot(xa_array,ya_array,'r-', label='Linha a')
ax.plot(xb_array,yb_array,'g-', label='Linha b')
ax.grid()

ax.set(xlabel='x', ylabel= 'y',title='Gráfico')
plt.legend()
```

```python
#Escala Log
from numpy.core.function_base import linspace
xc_array = np.linspace(0,10,500)
yc_array = np.exp(xc_array)

plt.figure()
plt.suptitle('Comparação escalas')

plt.subplot(211)
plt.plot(xc_array, yc_array,'r-')
plt.yscale('log')
plt.ylabel('y log')
plt.grid()

plt.subplot(212)
plt.plot(xc_array, yc_array,'k-')
plt.ylabel('y linear')
plt.grid()

plt.show()
```

```python
#Eixos secundarios
xc_array
yc_array

fig, ax1 =plt.subplots()
ax1.plot(xc_array, yc_array,'r-')
ax1.set_ylabel('Y1',color='r')

ax2 =ax1.twinx()
ax2.plot(xc_array,yc_array,'g-')
ax2.set_ylabel('y2',color='g')

plt.grid()
plt.yscale('log')
plt.show()
```
